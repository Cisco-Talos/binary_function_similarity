```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'

from itertools import product
```

```python
dataset = pd.read_csv("testing_Dataset-2.csv", index_col=0)
assert(dataset.isnull().values.any() == False)
print(dataset.shape)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15,15))
coordinates = list(product(list(range(3)), list(range(2))))
keys = ['library', 'arch', ['arch', 'bit'], 'compiler', ['compiler', 'version'], ['compiler', 'optimizations']]
for cord, key_v in zip(coordinates, keys):
    dataset.groupby(key_v).count()['idb_path'].sort_values(ascending=False).plot.bar(ax=axes[cord])
plt.subplots_adjust(hspace=0.9)
plt.show()
```
